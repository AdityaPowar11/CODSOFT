```python
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('advertising.csv')
df = pd.DataFrame(data)

print(df.info())
print(df.shape)
print(df.describe())

# Data Visualization
# Scatter plot for TV vs Sales
sns.scatterplot(x='TV', y='Sales', data=df)
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.show()

# Scatter plot for Radio vs Sales
sns.scatterplot(x='Radio', y='Sales', data=df)
plt.title('Radio Advertising vs Sales')
plt.xlabel('Radio Advertising Spend')
plt.ylabel('Sales')
plt.show()

# Scatter plot for Newspaper vs Sales
sns.scatterplot(x='Newspaper', y='Sales', data=df)
plt.title('Newspaper Advertising vs Sales')
plt.xlabel('Newspaper Advertising Spend')
plt.ylabel('Sales')
plt.show()

# Regression plot for TV vs Sales
sns.regplot(x='TV', y='Sales', data=df)
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.show()

# Regression plot for Radio vs Sales
sns.regplot(x='Radio', y='Sales', data=df)
plt.title('Radio Advertising vs Sales')
plt.xlabel('Radio Advertising Spend')
plt.ylabel('Sales')
plt.show()

# Regression plot for Newspaper vs Sales
sns.regplot(x='Newspaper', y='Sales', data=df)
plt.title('Newspaper Advertising vs Sales')
plt.xlabel('Newspaper Advertising Spend')
plt.ylabel('Sales')
plt.show()
```

```python
TV_sales = df.groupby('TV')["Sales"].mean()
df['TV_sales'] = df['TV'].map(TV_sales)
```

```python
Radio_sales = df.groupby('Radio')["Sales"].mean()
df['Radio_sales'] = df['Radio'].map(Radio_sales)
Newspaper_sales = df.groupby('Newspaper')["Sales"].mean()
df['Newspaper_sales'] = df['Newspaper'].map(Newspaper_sales)
```

```python
df.to_csv('Sales.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df.drop("Sales",axis=1)
Y=df["Sales"]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=52)

model=LinearRegression()
model.fit(X_train,Y_train)

X_test_prediction= model.predict(X_test)

mse =mean_squared_error(Y_test,X_test_prediction)
print(f"Mean Squared Error (MSE): {mse:.2f}")

r2 = r2_score(Y_test,X_test_prediction)
print(f"R-squared score: {r2:.2f}")
```
